### Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
train = transposed_lasVegas_data.loc[transposed_lasVegas_data.index < '2018-01-31']
test = transposed_lasVegas_data.loc[transposed_lasVegas_data.index >= '2018-01-31']
```

```python
transposed_lasVegas_data.isna().sum()
```

```python
transposed_lasVegas_data.columns
```

```python
# features and labels
features = ['Paradise', 'Spring Valley', 'Sunrise Manor', 'Enterprise',
       'Centennial Hills', 'Lone Mountain', 'Michael Way',
       'North Cheyenne', 'Charleston Heights']
label = 'average'
```

```python
# slicing
X_train  = train[features]
y_train = train[label]


X_test = test[features]
y_test  = test[label]
```

```python
X_test
```

```python
linear_model = LinearRegression()   # initialize the model
```

```python
linear_model.fit(X_train, y_train)
```

```python
pred = linear_model.predict(X_test)
```

```python
pred
```

```python
df_predict = pd.DataFrame(pred)
```

```python
df_predict.rename(columns = {0:"predicted_values"}, inplace=True)
```
